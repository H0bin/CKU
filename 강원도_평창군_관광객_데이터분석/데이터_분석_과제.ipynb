```python
file = "./data.csv"
import pandas as pd 
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

data = pd.read_csv(file, encoding='cp949')
data.head()
```

```python
for i in range(4, 16):
    data.iloc[:,i] = data.iloc[:,i].astype(str).str.replace(',', '')
    data.iloc[:,i] = data.iloc[:,i].astype(str).str.replace('-', '0')
data.head(9)
```

```python
data.iloc[:,3:16] = data.iloc[:,3:16].apply(pd.to_numeric)
data.info()
```

```python
df = data.iloc[:,:-1]
df.head(10)
```

```python
df["평균"] = df.iloc[:,4:16].mean(axis=1)
df['봄'] = df.iloc[:,6:9].sum(axis=1)
df['여름'] = df.iloc[:,9:12].sum(axis=1)
df['가을'] = df.iloc[:,12:15].sum(axis=1)
df['겨울'] = df.iloc[:,[15, 4, 5]].sum(axis=1)
df.head()
```

```python
#먼저 내국인 데이터 외국인 데이터 나누기. 
local = df[df.iloc[:,2] == '내국인']
foreigner = df[df.iloc[:,2] == '외국인']

print(local.head())
print(foreigner.head())
```

```python
local = local.reset_index(drop=True) #인덱스 초기화
foreigner = foreigner.reset_index(drop=True)
local
```

```python
# 2010 년도 ~ 2021 년도 각 년도마다 총합 방문률이 가장 높은 수, 가장 작은 수 
def find_top_bottom(df):
    year_top = []
    year_bottom = []
    year = 2010
    top = 0
    bottom = 1000000000
    alls = 0
    for i in range(len(df)):
        alls = df.iloc[i, 16]
        if year == df.iloc[i, 3]:
            if alls > top:
                top = alls
            if alls < bottom:
                
                bottom = alls
        else:
            year += 1
            year_top.append(top)
            year_bottom.append(bottom)
            top = df.iloc[i, -1]
            bottom = df.iloc[i, -1]

    year_top.append(top)
    year_bottom.append(bottom)

    return year_top, year_bottom
    
local_top, local_bottom = find_top_bottom(local)
foreigner_top, foreigner_bottom = find_top_bottom(foreigner)
```

```python
def sessons(df, sesson):
    top = 0
    bottom = 100000000
    year = 2010
    tops = []
    bottoms = []
    for i in range(len(df)):
        s = df.iloc[i, sesson]
        if year == df.iloc[i, 3]:
            if s > top :
                top = s
                
            if s < bottom:
                bottom = s
        else:
            year += 1
            tops.append(top)
            bottoms.append(bottom)
            top = s
            bottom = s
                
    tops.append(top)
    bottoms.append(bottom) 
        
    return tops, bottoms

l_spring_top, l_spring_bottom = sessons(local, 17)
l_summer_top, l_summer_bottom = sessons(local, 18)
l_autumn_top, l_autumn_bottom = sessons(local, 19)
l_winter_top, l_winter_bottom = sessons(local, 20)
f_spring_top, f_spring_bottom = sessons(foreigner, 17)
f_summer_top, f_summer_bottom = sessons(foreigner, 18)
f_autumn_top, f_autumn_bottom = sessons(foreigner, 19)
f_winter_top, f_winter_bottom = sessons(foreigner, 20)
```

```python
local_spring_top = local['봄'][local['봄'].isin(l_spring_top)].index
local_spring_bottom =  local['봄'][local['봄'].isin(l_spring_bottom)].index
local_summer_top = local['여름'][local['여름'].isin(l_summer_top)].index
local_summer_bottom =  local['여름'][local['여름'].isin(l_summer_bottom)].index
local_actumn_top = local['가을'][local['가을'].isin(l_autumn_top)].index
local_actumn_bottom =  local['가을'][local['가을'].isin(l_autumn_bottom)].index
local_winter_top = local['겨울'][local['겨울'].isin(l_winter_top)].index
local_winter_bottom =  local['겨울'][local['겨울'].isin(l_winter_bottom)].index

foreigner_spring_top = foreigner['봄'][foreigner['봄'].isin(f_spring_top)].index
foreigner_spring_bottom =  foreigner['봄'][foreigner['봄'].isin(f_spring_bottom)].index
foreigner_summer_top = foreigner['여름'][foreigner['여름'].isin(f_summer_top)].index
foreigner_summer_bottom =  foreigner['여름'][foreigner['여름'].isin(f_summer_bottom)].index
foreigner_actumn_top = foreigner['가을'][foreigner['가을'].isin(f_autumn_top)].index
foreigner_actumn_bottom =  foreigner['가을'][foreigner['가을'].isin(f_autumn_bottom)].index
foreigner_winter_top = foreigner['겨울'][foreigner['겨울'].isin(f_winter_top)].index
foreigner_winter_bottom =  foreigner['겨울'][foreigner['겨울'].isin(f_winter_bottom)].index
```
