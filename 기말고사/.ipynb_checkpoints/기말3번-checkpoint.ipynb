## 1단계 : 데이터 전처리

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('mode.chained_assignment',  None) # <==== 경고를 끈다

plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False

file = "./data/subwayfee.csv"
df = pd.read_csv(file, encoding='cp949')
df.tail() 
```

## 데이터 전처리

```python
df["사용월"].unique() # array(['2019-01'], dtype=object)
#사용월을 보니 2019년 1월 데이터이다.
```

```python
df["호선명"].unique() 
#array(['1호선', '2호선', '3호선', '4호선', '경부선', '경인선', '경원선', '안산선', '과천선',
#       '분당선', '일산선', '중앙선', '장항선', '경의선', '경춘선', '수인선', '경강선', '5호선',
#       '6호선', '7호선', '8호선', '9호선', '공항철도 1호선', '9호선2~3단계', '우이신설선'],
#      dtype=object)
```

```python
df.info() 
#인원 정보는 전부 int, 역id 또한 int 형이다.
#그리고 결측치도 없다.
#<class 'pandas.core.frame.DataFrame'>
#RangeIndex: 598 entries, 0 to 597
#Data columns (total 8 columns):
# #   Column  Non-Null Count  Dtype 
#---  ------  --------------  ----- 
# 0   사용월     598 non-null    object
# 1   호선명     598 non-null    object
# 2   역ID     598 non-null    int64 
# 3   지하철역    598 non-null    object
# 4   유임승차    598 non-null    int64 
# 5   유임하차    598 non-null    int64 
# 6   무임승차    598 non-null    int64 
# 7   무임하차    598 non-null    int64 
#dtypes: int64(5), object(3)
#memory usage: 37.5+ KB
```

```python
result = []
for i in range(len(df)): #0값이 들어간 데이터는 제외
    check = df.iloc[i, :]
    if (check[4] != 0) & (check[6] != 0) & (check[5] != 0) & (check[7] != 0):
        result.append(i)            
```

```python
dft = df.iloc[result, :]
dft.reset_index(inplace=True)
```

```python
dft["승차인원"] = dft["유임승차"] + dft["무임승차"]
dft["하차인원"] = dft["유임하차"] + dft["무임하차"]
```

```python
dft["유임승차비율"] = dft["유임승차"] / dft["승차인원"]
dft["무임승차비율"] = dft["무임승차"] / dft["승차인원"]
dft["유임하차비율"] = dft["유임하차"] / dft["하차인원"]
dft["무임하차비율"] = dft["무임하차"] / dft["하차인원"]
```

```python
def tops(df, target):
    top = 0
    idx = 0
    result = []
    for k in target:
        for i in range(len(df)):
            check = df[k][i]
            if check > top:
                top = check
                idx = i
        result.append(df.iloc[idx, :])
        top = 0
        idx = 0
    return result

def bottoms(df, target):
    bottom = 1000000000
    idx = 0
    result = []
    for k in target:
        for i in range(len(df)):
            check = df[k][i]
            if check < bottom:
                bottom = check
                idx = i
        result.append(df.iloc[idx, :])
        bottom = 1000000000
        idx = 0
        
    return result
```

```python
target = ["유임승차비율", "무임승차비율", "유임하차비율", "무임하차비율"]
top = tops(dft, target)
```

## 데이터 분석

```python
for i in range(len(top)):
    print(target[i]+"이 가장 높은 역은 "+top[i]["지하철역"]+"이고", target[i]+"은", round(top[i][target[i]], 3), "이다.")
#유임승차비율이 가장 높은 역은 홍대입구이고 유임승차비율은 0.953 이다.
#무임승차비율이 가장 높은 역은 소요산이고 무임승차비율은 0.563 이다.
#유임하차비율이 가장 높은 역은 홍대입구이고 유임하차비율은 0.957 이다.
#무임하차비율이 가장 높은 역은 소요산이고 무임하차비율은 0.596 이다.
```

> 어떤 역이 특별한 값을 가지는지 확인

```python
top_alls = ["승차인원", "하차인원"]
top_all = tops(dft, top_alls) #어떤 역이 가장 많은지 알아보기
bottom_all = bottoms(dft, top_alls) #어떤 역이 가장 적은지 알아보기
```

```python
for i in range(len(top_all)):
    print(top_alls[i]+"이 가장 많은 역은 "+top_all[i]["지하철역"]+"역이고", top_alls[i]+"은", round(top_all[i][top_alls[i]], 3),"명이다.")
    print(top_alls[i]+"이 가장 적은 역은 "+bottom_all[i]["지하철역"]+"역이고", top_alls[i]+"은", round(bottom_all[i][top_alls[i]], 3),"명이다.")
    print("\n")

#승차인원이 가장 많은 역은 강남역이고 승차인원은 3339904 명이다.
#승차인원이 가장 적은 역은 지평역이고 승차인원은 1116 명이다.


#하차인원이 가장 많은 역은 강남역이고 하차인원은 3378103 명이다.
#하차인원이 가장 적은 역은 옥수역이고 하차인원은 879 명이다.
```
